```python
from PIL import Image
import os

# Define the directories
extract_to = 'D:\Dropbox\PC (2)\Desktop\extract'  # Directory to extract the contents
source_directory = extract_to   # Directory with extracted images
target_directory = 'D:\Dropbox\PC (2)\Desktop\converted'  # Directory to save converted images

# Create the target directory if it doesn't exist
os.makedirs(target_directory, exist_ok=True)

def convert_images_to_jpg(source_dir, target_dir):
    for root, _, files in os.walk(source_dir):
        for file in files:
            file_path = os.path.join(root, file)
            try:
                # Construct the new file path with .jpg extension
                new_file_path = os.path.join(target_dir, os.path.splitext(file)[0] + '.jpg')
                with Image.open(file_path) as img:
                    img.save(new_file_path, 'JPEG')
                    # img.convert('RGB').save(new_file_path, 'JPEG')
                    print(f"Converted {file} to {new_file_path}")
            except Exception as e:
                print(f"Error converting {file_path}: {e}")

# Convert the images
convert_images_to_jpg(source_directory, target_directory)

# Verify the conversion
print(f"Contents of {target_directory}:")
for root, dirs, files in os.walk(target_directory):
    for name in files:
        print(os.path.join(root, name))
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define the parameters
img_width, img_height = 300, 300  # Adjust the target size as needed
batch_size = 32

# Image Data Generator for preprocessing
datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

# Define the data directory
data_dir = target_directory  # Directory with converted images

# Training data generator
train_generator = datagen.flow_from_directory(
    data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary',
    subset='training'
)

# Validation data generator
validation_generator = datagen.flow_from_directory(
    data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary',
    subset='validation'
)

# Print the number of images found
print(f"Training images found: {train_generator.samples}")
print(f"Validation images found: {validation_generator.samples}")
```
